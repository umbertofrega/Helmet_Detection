# Helmet and Vest Detection 

The problem falls in the bigger family of the *Object Detection* 

We start by importing the model of choice YOLO and the packages for showing the results and doing some analysis

```python
from ultralytics import YOLO
```

Then we import the dataset

```python
source ="dataset/1.mp4"
```

```python
model = YOLO("model/yolov8s.pt")

model.train()
```

```python
cap = cv2.VideoCapture(source)

while cap.isOpened():
    success, frame = cap.read()

    if success:
        results = model(frame)

        annotated_frame = results[0].plot()

        cv2.imshow("YOLO Inference", annotated_frame)

        if cv2.waitKey(1) & 0xFF == ord("q"):
            break
    else:
        break

cap.release()
cv2.destroyAllWindows()
```
